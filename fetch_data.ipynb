```python
indexes=[]
with open("datav3.json","a") as f:    
    cnt=0
    while cnt<=10000:
        #time.sleep(1)
        index=random.randint(1,8000000)
        #print(index)
        if index not in indexes:
            try:
                book = gc.book(index)
                if int(book.ratings_count)>=1:
                    indexes.append(index)
                    #print(cnt)
                    tempd={}
                    tempd['ID']=index
                    tempd['title']=book.title
                    tempd['num_pages']=book.num_pages
                    tempd['author']=book.authors[0].name
                    tempd['rating']=book.average_rating
                    tempd['ratings_count']=book.ratings_count
                    tempd['text_reviews_count']=book.text_reviews_count
                    tempd['language']=book.language_code
                    js=json.dumps(tempd)
                    f.write(js)
                    cnt+=1
                    print("cnt:%s"%cnt)
                
                
                else:
                    print('%s not enough ratings'%index)
                    continue
            except:
                print('%s err'%index)
                continue
                
```
